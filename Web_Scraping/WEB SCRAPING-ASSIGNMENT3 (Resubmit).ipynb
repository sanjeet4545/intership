### 2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. 

***In case if any product has less than 3 pages in search results thenscrape all the products available under that product name. Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“. ***

```python
# creating an empty list for scrap data
Brand = []
Product = []
Price = []
Return = []
Delivery = []
Availability = []
URL = []
```

```python
URL = []
start = 0
end = 2
for page in range(start,end):
    url = driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url:
        URL.append(i.get_attribute('href'))
    time.sleep(2)
    next_button = driver.find_elements(By.XPATH,'//a[@aria-label="Go to next page, page 2"]')
```

```python
len(URL)
```

```python
# iterating each and every url to fectch
from tqdm import tqdm
for i in tqdm(URL):
    driver.get(i)
    time.sleep(3)

    # Extracting Brand
    try:
        brand_tag = driver.find_element(By.XPATH,'//a[@id="bylineInfo"]')
        Brand.append(brand_tag.text)
    except NoSuchElementException:
        Brand.append('-')
        
    time.sleep(1)
    
    # Extracting Product
    try:
        product_tag = driver.find_element(By.XPATH,'//span[@id="productTitle"]')
        Product.append(product_tag.text)
    except NoSuchElementException:
        Product.append('-')
        
    time.sleep(1)
        
    # Extracting Price
    
    try:
        price_tag = driver.find_element(By.XPATH,'//span[@class="a-price-whole"]')
        Price.append(price_tag.text)
    except NoSuchElementException:
        Price.append('-')
        
    time.sleep(1)
    
    # Extracting Return
    
    try:
        return_tag = driver.find_element(By.XPATH,'//*[@id="RETURNS_POLICY"]/span/div[2]')
        Return.append(return_tag.text)
    except NoSuchElementException:
        Return.append('-')
        
    time.sleep(1)
    
    # Extracting Delivery
    try: 
        delivery_tag = driver.find_element(By.XPATH,'//div[@id="deliveryBlockMessage"]')
        Delivery.append(delivery_tag.text)
    except NoSuchElementExcpetion:
        Delivery.append('-')
    
    time.sleep(1)
    
    # Extracting Availability
    try:
        availability_tag = driver.find_element(By.XPATH,'//span[@class="a-size-medium a-color-success"]')
        Availability.append(availability_tag.text)
    except NoSuchElementException:
        Availability.append('-')
        
    time.sleep(1)
    
    driver.close()
```

```python
print(len(Brand),len(Product),len(Price),len(Return),len(Delivery),len(Availability),len(URL))
```

### 5. Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps. 

```python
# Connecting the webdriver
driver = webdriver.Chrome()

# Open the url in google chrome
driver.get('https://www.google.co.in/maps')
user_input=input("Enter city name for which coordinates is required : ")
time.sleep(5)

# Search bar
search = driver.find_element(By.XPATH,'//input[@jslog="11886"]')
search.send_keys(user_input)

# Search Button
search_button = driver.find_element(By.XPATH,'//button[@jslog="11887"]')
search_button.click()
time.sleep(5)

lat = []
long = []

try: 
    url_string = driver.current_url
    print("URL Extracted",url_string)
    lat_long = re.findall(r'@(.*)data',url_string)   
    if len(lat_long):
        lat_long_list = lat_long[0].split(",")
        if len(lat_long_list)>=2:
            lat.append(lat_long_list[0])
            long.append(lat_long_list[1])
except Exception as e:
    print("error:",str(e))

time.sleep(2)
driver.close()

location_data=pd.DataFrame({'Enter Location':user_input,'latitude':lat,'longitude':long})
location_data
```

```python
# Connecting to webdriver
driver = webdriver.Chrome()

try:
    # Opening url in chrome browser
    driver.get("https://www.hostelworld.com/")

    loc = driver.find_element(
        By.XPATH,
        "/html/body/div[3]/div/div/div[1]/div[1]/div/div[2]/div[4]/div/div[2]/div/div[1]/div/div/div/input",
    )

    loc.send_keys("london")
    search_btn = driver.find_element(
        By.XPATH,
        "/html/body/div[3]/div/div/div[2]/div[1]/div/div/div[4]/div/div[2]/div/div[5]/button",
    )
    search_btn.click()

except Exception as e:

    driver.get(
        "https://www.hostelworld.com/s?q=London, England&country=England&city=London&type=city&id=3&from=2022-07-28&to=2022-07-31&guests=2&HostelNumber"
    )

time.sleep(3)

#lets find required data
Hostel_name = []
Distance = []
Rating = []
Total_reviews = []
Overall_Review = []
Privates_price = []
Dorms_price = []
Facilities = []
Description =[]
Product_url = []
```

```python
# Fetching distance
try:
    dist = driver.find_elements(By.XPATH,'//span[@class="distance-description"]')
    for i in dist:
        Distance.append(i.text.replace('Hostel\n-',' '))
except NoSuchElementException:
    Distance.append('-')
      
# Fetching Privates_price
try:
    privates = driver.find_elements(By.XPATH,'//div[@class="property-accommodation-price"][1]')
    for i in privates:
        Privates_price.append(i.text.split('%')[-1].replace('\n',''))
         
except NoSuchElementException:
    Privates_price.append('-')
       
# Fetching Dorms_price
try:
    dorms = driver.find_elements(By.XPATH,'//div[@class="property-accommodation-price"][2]')
    for i in dorms:
        Dorms_price.append(i.text.split('%')[-1].replace('\n',''))
except NoSuchElementException:
    Dorms_price.append('-')
    
# URL
url = driver.find_elements(By.XPATH,'//a[@class="property-card-container"]') 
for i in url:
    Product_url.append(i.get_attribute("href"))
    
for i in Product_url:
    driver.get(i)
    time.sleep(5)
                  
# Fetching hotel name
    try:
        name = driver.find_elements(By.XPATH,'//div[@class="title-2"]')
        for i in name:
            Hostel_name.append(i.text)
    except NoSuchElementException:
        Hostel_name.append('-')
         
# Fetching Rating
    try:
        rating = driver.find_elements(By.XPATH,'//div[@class="flex-20"]/div[1]/div[1]')
        for i in rating:
            Rating.append(i.text)
    except NoSuchElementException:
        Rating.append('-')
        
        
# Fetching Total_reviews
    try:
        review = driver.find_elements(By.XPATH,'//div[@class="flex-20"]/div[1]/div[2]/div[2]')
        for i in review:
            Total_reviews.append(i.text)
    except NoSuchElementException:
        Total_reviews.append('-')
        
# Fetching Overall_Review
    try:
        overall = driver.find_elements(By.XPATH,'//div[@class="flex-20"]/div[1]/div[2]/div[1]')
        for i in overall:
            Overall_Review.append(i.text)
    except NoSuchElementException:
        Overall_Review.append('-')
        
# Fetching Facilities                                     
    try:
        facilities = driver.find_elements(By.XPATH,'//ul[@class="facility-sections"]/li[1]/ul[1]')
        for i in facilities:
            Facilities.append(i.text)

    except NoSuchElementException:
        Facilities.append('-')
               
# Fetching Description     
    try:
        Desc = driver.find_elements(By.XPATH,'//div[@class="body-3"]')
        for i in Desc:
            Description.append(i.text)
    except NoSuchElementException:
        Description.append('-')
        
driver.close()
```
